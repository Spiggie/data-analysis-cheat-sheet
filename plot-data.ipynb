# Plot Data

## Imports
`%matplotlib inline` let's us see plots in the notebook directly.

```python
import pandas as pd
% matplotlib inline
```

```python
# read the file
df = pd.read_csv('data\earthquakes.csv')
df.head()
```

```python
# show histograms for numerical data
df.hist(figsize=(8, 8));
```

```python
# show a histogram for a special column
df['mag'].hist()
```

```python
# this can also be done like this
df['mag'].plot(kind='hist');
```

```python
# get quick insights of numerical data
pd.plotting.scatter_matrix(df, figsize=(15, 15));
```

```python
# plot chosen x and y axis with a scatter plot
df.plot(x='longitude', y='latitude', kind='scatter');
```

```python
# box plot
df['mag'].plot(kind='box')
```
